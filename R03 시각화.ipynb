### 그래프 설정
+ main : 제목

+ xlab/ylab : 축제목
+ axes : 축 눈금

### 그래프 눈금 세부조정
+ axes 함수 이용
+ axis 함수 이용
+ axis(축지정, 눈금지정)

```r
plot(heights, main='studens heights', sub='class A', xlab='numbers', col='blue', pch=3, axes=F)
axis(side=1, at=1:10)
axis(side=2, at=seq(140,185,5))
grid()
text(heights)
```

```r
axis(side=1, at=1:10) # x축 설정
```

```r
axis(side=2, at=seq(140,185,5)) #y축 설정
```

### 그래프 텍스트 추가하기
+ text(y좌표, 글자크기, 글자위치, 글자내용
+ 특정 x,y위치에 글자 표시
+ adj : 음수 - 오른쪽 위, 양수 - 왼쪽 아래

```r
plot(heights, col='red')
text(heights, cex=1.0, adj=c(0, -0.2))
grid()
```

```r
plot(dose, drugA, main='drugpower', sub='addict', xlab='age', ylab='effect', col='red', pch=3, axes=F)
text(dose, drugA, cex = 1.0, adj=c(0, -0.5), lab=drugA)
axis(side=1, at=seq(10,80,5))
axis(side=2, at=seq(10,70,5))
box()
grid()
```

### 범주형 데이터를 토대로 붓꽃의 품종을 각기 다른 기호, 색상으로 표현

```r
plot(iris$Petal.Length, iris$Petal.Width, xlab='length', ylab='width', pch=as.integer(iris$Species))
```

```r
plot(iris$Petal.Length, iris$Petal.Width, xlab='length', ylab='width', col=as.integer(iris$Species)+1)
```

```r
plot(iris$Petal.Length, iris$Petal.Width, xlab='length', ylab='width', col=as.integer(iris$Species)+1, pch=as.integer(iris$Species))
```

### 범례 표시하기
+ legend(위치, 값, 기호, 색상)
+ 위치: top, bottom, right, left, center

```r
plot(iris$Petal.Length, iris$Petal.Width, xlab='length', ylab='width', col=as.integer(iris$Species)+1, pch=as.integer(iris$Species))
legend('topleft', c('setosa','versicolor','verginica'), pch=c('○','△','♧'), col=c('red','green','blue'))
```

### 자동으로 텍스트 기호 색상 지정

```r
plot(iris$Petal.Length, iris$Petal.Width, xlab='length', ylab='width', col=as.integer(iris$Species)+1, pch=as.integer(iris$Species))
legend('bottomright', levels(iris$Species), pch=unique(as.integer(iris$Species)), col=unique(as.integer(iris$Species)+1))
```

```r
manager <- c(1, 2, 3, 4, 5)
country <- c('US', 'US', 'UK', 'UK', 'UK')
gender <- c('M', 'F', 'F', 'M', 'F')
age <- c(32, 45, 25, 39, 99)
```

### 벡터들로 데이터프레임 생성하기
+ data.frame(벡터1, 벡터2, 벡터3...)

```r
managers <- data.frame(manager, country, gender, age)
```

```r
str(managers)
summary(managers)
```

```r
mcountry <- managers$country
mage <- managers$age
mgender <- managers$gender
```

```r
plot(mcountry, mage)
plot(mgender, mage)
```

```r
plot(mage, pch=unique(as.integer(mcountry)), col=))
legend('topleft', levels(mcountry), pch=unique(as.integer(mcountry)), col=c('red', 'green'))
```

```r
plot(mage, pch=as.integer(mgender), col=as.integer(mgender)+1)
legend('topleft', levels(mgender), pch=unique(as.integer(mgender)), col=c('red', 'green'))
```
