### 두 표본 가설검정

+ 단일표본의 비교문제는 하나의 모집단에서 추출된 하나의 표본에 대해 검정
+ 그에 비해 두 표본 비교문제는 독립적인 두 모집단으로 부터 각각 추출된 표본의 경우, 동일한 모집단으로부터 추출된 두 표본의 경우로 나누어 검정을 실시함
    - 남녀 소득차이 비교(독립표본)
    - 과외 받기전 받기후 성적비교(짝 표본)
+ => 2개의 집단간의 차이가 의미 있는지 확인
+ 일상에서는 단일 표본문제보다 더 많이 사용됨

### 정규성 검정

+ 데이터의 분포가 정규분포를 따르는지 알아보는 것을 의미

+ 통계의 다양한 검정법은 데이터의 정규분포를 가정하고 수행됨

+ 데이터가 충분히 크다면(보통 30개 이상) 중심극한정리를 적용할 수 있음

+ 하지만, 그 이하일 경우 별도의 정규성 검정이 필요할 수 있음

+ 독립표본을 이용해서 가설검정을 실시하기 위해서는 반드시 **정규성**, **등분산성**이 만족되어야 함



```r
x = c(175,168,168,190,156,181,182,175,174,179)

y = c(185,169,173,173,188,186,175,174,179,180)
```

```r
hist(x, probability=T)
lines(density(x), col='red')
hist(y, probability=T)
lines(density(y), col='red')
```

```r
qqnorm(x)
qqline(x, col='red')
qqnorm(y)
qqline(y, col='red')
```

```r
shapiro.test(x)
shapiro.test(y)
```

```r
boxplot(x, y, names=c('x','y'))
```

```r
# ex) 잔디 깍기 기계 생산 공장에서 작업시간 단축을 위해 2가지 공정이 제안되었음

# A 공정방식은 5명의 근로자에 의해, B 공정방식은 6명의 근로자에 의해

# 각각 테스트가 진행되었을때, 작업시간의 차이가 있는지 검정하시오.

# 유의수준 0.10을 사용한다

# A 공정 : 2 4 9 3 2 (단위: 분)

# B 공정 : 3 7 5 8 4 3
```

```r
A <- c(2,4,9,3,2)
B <- c(3,7,5,8,4,3)
```

```r
shapiro.test(A)
shapiro.test(B)
```

### 독립표본 가설검정
+ 서로 다른 두 표본간의 평균의 차이가 유의미한지 판단이 목적
+ t.test(표본1, 표본2, 등분산여부, 신뢰범위, 검정유형)

```r
# 귀무가설 : 공정방식에 따른 작업시간의 차이가 없다
# 대립가설 : 공정방식에 따른 작업시간의 차이가 있다
# 방향성이 없으므로 양측검정
# 자유도 : 독립표본 검정시 (n1-1) + (n2-1) = 9 (단, 등분산성을 만족할 때의 계산식. 만족하지 않을 경우 합에서 -1)
# 임계값 : 1.833, -1.833
```

```r
t.test(A, B, alternative=c('two.sided')) # var.equal = F(등분산성 기본값은 F)
```

### 표본간 등분산성을 확인하려면 var.test함수를 사용함
+ var.test(표본1, 표본2)
+ 귀무가설 : 두 표본간의 분산 차이가 없음 
+ 대립가설 : 두 표본간의 분산 차이가 있음
+ 유의수준 : 0.05

```r
var.test(A, B)
```

```r
t.test(A, B, alternative=c('two.sided'), var.equal=T)
```

```r
sales <- c(131, 135, 146, 165, 136, 142)
audits <- c(130, 102, 129, 143, 149, 120, 139)
```

```r
# 영업사원과 감사부사원의 하루 출장비에 차이가 있는지 검정하시오 (유의수준 0.10)

# 영업부서가 감사부서보다 출장비가 많다고 할 수 있는가?
```

```r
# 귀무가설 : 영업부서의 출장비는 감사부서보다 많지 않다
# 대립가설 : 영업부서의 출장비는 감서부서보다 많다
```

```r
shapiro.test(sales)
shapiro.test(audits) # 정규성 만족
```

```r
var.test(sales, audits) # 등분산성 만족
```

```r
t.test(sales, audits, alternative=('greater'), var.equal=T)
```

```r
# p밸류 0.0763이 유의수준 0.10 안에 있고 t통계량 1.5368이 임계값 1.36보다 바깥에 있으므로 기각
# 영업 부서가 더 많이 받는다
```

```r
# 모델회사의 두 모델 수입이 다음과 같을때

# kim이라는 모델이 park 이라는 모델의 수입보다 크다고 할수 있는지 유의수준 0.05로 검정하시오

# kim : 5, 4.5, 3.4, 3.4, 6, 3.3, 4.5, 4.6, 3.5, 5.2, 4.8, 4.4, 4.6, 3.6, 5.0

# park : 3.1, 3.7, 3.6, 4.0, 3.8, 3.8, 5.9, 4.9, 3.6, 3.6, 2.3, 4.0
```

```r
kim <- c(5, 4.5, 3.4, 3.4, 6, 3.3, 4.5, 4.6, 3.5, 5.2, 4.8, 4.4, 4.6, 3.6, 5.0)
park <- c(3.1, 3.7, 3.6, 4.0, 3.8, 3.8, 5.9, 4.9, 3.6, 3.6, 2.3, 4.0)
```

```r
shapiro.test(kim)
shapiro.test(park)
```

```r
var.test(kim, park)
```

```r
t.test(kim, park, alternative=('greater'), var.equal=T)
```

```r
# 임계값 : 1.708
# t통계량 1.6367이 임계값보다 안에 있고 p밸류가 유의수준0.05보다 바깥에 있으므로 귀무가설 기각, 대립가설 채택
# 김이 박보다 평균 모델수입이 크지 않을 가능성이 높다
```

```r
hist(kim, probability = T)
lines(density(kim), col='red')
hist(park, probability = T)
lines(density(park), col='red')
```
