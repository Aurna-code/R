### 작업한 데이터 저장하기
+ R에서 만든 객체(벡터, 데이터프레임)들은 메모리 내부에 생성
+ 프로그램이 종료하면 메모리 내부 객체들은 사라짐
+ 메모리 내 생성된 객체를 파일형태로 저장할 필요성 - 직렬화
+ 파일저장 : save(객체명, 저장경로)
+ 파일읽기 : load(파일경로)

```r
load('heights.rdata')
```

## 막대그래프
+ 비율척도상의 크기를 막대형태로 나타낸 것
    - 등간척도를 이용한 도수분포표를 막대그래프로 구현 가능
+ 주로 자료의 범주별 빈도를 요약해서 나타냄
+ barplot(대상, 누적여부, 표시방향)

```r
barplot(heights)
```

### 막대그래프 시각화
+ names.arg : x축 레이블 표시, 대상은 벡터로 정의

```r
barplot(heights, names.arg=c(1:10), col=rainbow(10))
```

### horiz : 세로/가로 막대그래프 여부 지정

```r
barplot(heights, names.arg=c(1:10), col=rainbow(10), horiz=T)
```

### text함수로 막대그래프에 숫자 표시

```r
bp <- barplot(heights, names.arg=c(1:10), col=rainbow(10), ylim=c(0,190), axes=F)
axis(side=2, at=seq(0, 200, 10))
text(bp, heights+5, lab=heights)
```

```r
bp <- barplot(heights, names.arg=c(1:10), col=rainbow(10), ylim=c(0,200), axes=F)
axis(side=2, at=seq(0, 200, 10))
text(bp, heights+5, lab=heights)
```

### mtcars 데이터셋 확인

```r
head(mtcars)
```

```r
mtcars$vs <- as.factor(mtcars$vs) #기어의 종류
mtcars$gear <- as.factor(mtcars$gear) # 기어수
```

#### 자동차 성능 지표 데이터에서 기어수gear, 기통수vs 시각화

```r
summary(mtcars)
```

```r
barplot(table(mtcars$gear))
```

### 누적막대 그래프 : stacked bar plot
+ 주 범주와 하위 범주간에 존재하는 양적 관계를 보여줌
+ 범주별 데이터를 비교하여 각 범주의 차이를 한눈에 파악하도록 시각화

#### 엔진종류별 기어수 시각화 : groupby

```r
counts <-table(mtcars$vs, mtcars$gear)
barplot(counts, col=c('orange', 'darkblue'), legend=c('v-shape', 'straight'))
```

### 그룹막대 그래프 : grouped bar plot

```r
counts <-table(mtcars$vs, mtcars$gear)
barplot(counts, col=c('orange', 'darkblue'), legend=c('v-shape', 'straight'), beside=T)
```

```r
titan = read.csv('titanic.csv')
```

```r
str(titan)
```

```r
titan$pclass <- as.factor(titan$pclass)
```

```r
str(titan)
```

```r
ss <-table(titan$sex, titan$survived)
bp2 <- barplot(ss, main='성별 생존현황', ylim=c(0,800), col=c('orange', 'Midnight blue'), names.arg=c('사망', '생존'), legend=c('female', 'male'), beside=T)
text(bp2, ss+20, lab=ss)
```

```r
ps <-table(titan$pclass, titan$survived)
bp3<- barplot(ps, main='좌석별 생존현황', ylim=c(0,600), col=c('orange', 'snow2', 'Midnight blue'), names.arg=c('사망', '생존'), legend=c('1','2','3'), beside=T)
text(bp3, ps+20, lab=ps)
```

```r
es <-table(titan$embarked, titan$survived)
bp1 <- barplot(es, main='승선위치별 생존현황', ylim=c(0,700), col=c('orange', 'snow2', 'Midnight blue'), names.arg=c('사망', '생존'), legend=c('C', 'Q', 'S'), beside=T)
text(bp1, es+20, lab=es)
```
